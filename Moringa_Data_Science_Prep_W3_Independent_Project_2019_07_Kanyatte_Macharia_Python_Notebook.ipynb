```
#This imports the pandas library
import pandas as pd
#This imports the numpy library
import numpy as np
```

```
#This imports the csv file from the url 'http://bit.ly/Telcom_dataset1' and loads it into a dataframe 
CDR_1 = pd.read_csv('http://bit.ly/Telcom_dataset1')
CDR_1_df = pd.DataFrame(CDR_1)
CDR_1_df
```

```
#This imports the csv file from the url 'http://bit.ly/Telcom_dataset2' and loads it into a dataframe 
CDR_2 = pd.read_csv('http://bit.ly/Telcom_dataset2')
CDR_2_df = pd.DataFrame(CDR_2)
CDR_2_df
```

```
#This imports the csv file from the url 'http://bit.ly/Telcom_dataset3' and loads it into a dataframe 
CDR_3 = pd.read_csv('http://bit.ly/Telcom_dataset3')
CDR_3_df = pd.DataFrame(CDR_3)
CDR_3_df
```

```
#This loads the cells_geo_description.xlsx file and then loads the first sheet of the file into a dataframe
cells_geo_des = pd.read_excel('cells_geo_description.xlsx')
cells_geo_des_df = pd.DataFrame(cells_geo_des)
cells_geo_des_df
```

```
#This loads the cells_geo_description.xlsx file and then loads the first sheet of the file into a dataframe
CDR_des = pd.read_excel('CDR_description.xlsx')
CDR_des_df = pd.DataFrame(CDR_des)
CDR_des_df
```

```
#This imports cells_geo_.csv file then loads it into a dataframe
cells_geo = pd.read_csv('cells_geo.csv', delimiter=';')
cells_geo_df = pd.DataFrame(cells_geo)
cells_geo_df
```

```
# This drops the columns: 'DECOUPZONE', 'LONGITUDE', 'LATITUDE' and 'AREA' from the dataframe: cells_geo_df
cells_geo_df.drop(['DECOUPZONE', 'LONGITUDE', 'LATITUDE', 'AREA'], axis=1, inplace=True)
cells_geo_df
```

```
#This renames the columns: 'PRODUTC' and 'DATETIME' to 'PRODUCT' and 'DATE_TIME' respectively in the
#dataframe: CDR_1_df
CDR_1_df.rename(columns={'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}, inplace=True)
CDR_1_df
```

```
#This renames the columns: 'DW_A_NUMBER' and 'DW_B_NUMBER' to 'DW_A_NUMBER_INT' and 'DW_B_NUMBER_INT' respectively
#in the dataframe: CDR_2_df
CDR_2_df.rename(columns={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'}, inplace=True)
CDR_2_df
```

```
#This renames the columns: 'CELLID' and 'SIET_ID' to 'CELL_ID' and 'SITE_ID' respectively
#in the dataframe: CDR_3_df
CDR_3_df.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID': 'SITE_ID'}, inplace=True)
CDR_3_df
```

```
#This concatenates the three dataframes: CDR_1_df, CDR_2_df and CDR_3_df to become one dataframe: CDR_final
CDR_final = pd.concat([CDR_1_df, CDR_2_df,CDR_3_df], ignore_index=True)
CDR_final
```

```
#This describes the value column in the dataframe: CDR_final 
CDR_final.describe()
```

```
#This code describes the value of various products in the CDR_final dataframe
CDR_final['VALUE'].groupby(CDR_final['PRODUCT']).describe()
```

```
#This code will give us the phone numbers of those those who implemented the sms and  product so as to upgrade those 
#products since they are way below the mean billing price
SmsPhoneNumber = CDR_final['DW_A_NUMBER_INT'].groupby(CDR_final['PRODUCT']).describe()
SmsPhoneNumber
```
