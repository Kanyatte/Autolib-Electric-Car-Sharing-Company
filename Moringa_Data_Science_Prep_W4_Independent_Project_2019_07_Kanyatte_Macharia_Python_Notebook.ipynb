```
#This imports the pandas library
import pandas as pd
#This imports the numpy library
import numpy as np
```

```
#This will read the link 'http://bit.ly/autolib_dataset' as a csv file and later load it into a datatframe
url = 'http://bit.ly/autolib_dataset'
autolib = pd.read_csv(url, delimiter=',')
autolib_df = pd.DataFrame(autolib)
autolib_df
```

```
#This code acceses the information in our dataset with regard to the number of values that are not null in the columns as well as the
#datatype for each column 
autolib_df.info()
```

```
#This drops some of the columns that are not required in analysis, columns that are similar to each other and columns which are 
#similar in all records i.e The year and month are 2018 and 4 throughout
autolib_df.drop(["Displayed comment", "Scheduled at", "ID", "Geo point", "Address", "year", "month"], axis=1, inplace=True)
autolib_df
```

```
#This renames the columns which had white spaces between words by replacing the white spaces with an underscore
autolib_df.rename(columns={'Bluecar counter': 'Bluecar_counter', 'Utilib counter': 'Utilib_counter', 'Utilib 1.4 counter': 'Utilib_1.4_counter', 'Charge Slots': 'Charge_Slots', 'Charging Status': 'Charging_Status', 'Postal code': 'Postal_code', 'Public name': 'Public_name', 'Rental status': 'Rental_status', 'Station type': 'Station_type', 'Subscription status': 'Subscription_status'}, inplace=True)
autolib_df
```

```
#This code is run to ensure the data is complete by checking for any null values that may be present in each column
autolib_df.isnull().any()
```

```
#This is to drop any duplicates that may be present in the data to improve its consistency
autolib_df.drop_duplicates()
autolib_df
```

```
#This will sort the data by the 'Public name' then by 'day' followed by 'hour' and finally 'minute'. This enables one to track 
#what happened in a station within the first nine days of April i.e. whether a car was picked, returned or nothing happened
# A car is returned if the value of the counter increases from the initial value to the next in the same station
# A car is rented out/picked if the value of the counter decreases from the initial value to the next in the same station
auto_sort = autolib_df.sort_values(by=['Public_name', 'day', 'hour', 'minute'],ascending=[True,True,True,True])
auto_sort
```

```
#This code groups the cumulative difference by the station so as to know if a car is picked, returned or nothing happened
#If the cumulative difference is '0' or 'Nan', nothing happened
#If it is less than 0, the negative number displays the number of cars rented out
#If it is greater than 0, the positive number shows the the number of cars returned to the station
auto_sort['Cum_Diff_Bluecar'] = auto_sort['Bluecar_counter'].groupby(auto_sort['Public_name']).diff(1)
auto_sort['Cum_Diff_Utilib'] = auto_sort['Utilib_counter'].groupby(auto_sort['Public_name']).diff(1)
auto_sort['Cum_Diff_Utilib_1.4'] = auto_sort['Utilib_1.4_counter'].groupby(auto_sort['Public_name']).diff(1)
auto_sort
```

```
#This code will replace the Nan values in the 'Cum_Diff_Bluecar', 'Cum_Diff_Utilib' and 'Cum_Diff_Utilib_1.4' columns  with '0.0'
auto_sort['Cum_Diff_Bluecar'].fillna(0.0, inplace=True)
auto_sort['Cum_Diff_Utilib'].fillna(0.0, inplace=True)
auto_sort['Cum_Diff_Utilib_1.4'].fillna(0.0, inplace=True)
auto_sort
```

```
#This code identifies the most popular hour of the day for picking a bluecar in Paris over April 2018
auto_sort[(auto_sort['City'] == 'Paris') & (auto_sort['Cum_Diff_Bluecar']<0)]['hour'].value_counts().idxmax()
```

```
#This code indentifies most popular hour for returning cars
auto_sort[(auto_sort['Cum_Diff_Bluecar']>0)]['hour'].value_counts().idxmax()
```

```
#This code identifies the most popular station in the entire company
most_popular_station = auto_sort['Public_name'].value_counts().idxmax()
most_popular_station
```

```
#This code identifies the most popular station at the most popular picking hour
auto_sort[(auto_sort['hour'] == 21)]['Public_name'].value_counts().idxmax()
```

```
#This identifies the most popular postal code for picking up blue cars
most_popular_code_for_picking_blue = auto_sort[(auto_sort['Cum_Diff_Bluecar']<0)]['Postal_code'].value_counts().idxmax()
most_popular_code_for_picking_blue
```

```
#This code determines whether the most popular station belongs to the most popular postal code for picking blue cars
if 'Public_name' == most_popular_station and 'Postal_Code' == most_popular_code_for_picking_blue:
  print("The most popular station belongs to the most popular postal code for picking blue cars")
else:
  print("The most popular station does not belong to the most popular postal code for picking blue cars")
```
